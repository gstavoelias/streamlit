```python
from server import Server
import time 

servidor = Server("http://127.0.0.1:8087/api/v1.0/")
for idx, row in df.iterrows():
    try:
        operador_id = row["TECNICO_ID"]
        rft_id = idx + 1
        solucao_id = row["SOLUCAO_ID"]
        horario = row["DATA"].strftime("%Y-%m-%d %H:%M:%S")
        duracao = 180

        print(f"[{idx}] Enviando para o banco: RFT={rft_id}, OPERADOR={operador_id}, SOLUCAO={solucao_id}, HORARIO={horario}")
        servidor.post_manutencao(operador_id=operador_id, rft_id=rft_id, solucao_id=solucao_id, horario=horario, duracao=duracao)
        time.sleep(0.5)
    except Exception as e:
        print(f"⚠️ Erro no índice {idx}: {e}")
```
